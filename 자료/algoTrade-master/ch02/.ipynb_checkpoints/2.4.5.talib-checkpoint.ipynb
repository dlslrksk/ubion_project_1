```python
import talib
import numpy as np
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pprint(talib.get_function_groups())
```

```python
# list of functions
talib.get_functions()
```

```python
np.random.seed(0)
close = np.random.random(250)
sma = talib.SMA(close, timeperiod=10)
plt.figure(figsize=(11,3))
plt.plot(close, "r-")
plt.plot(sma, "b-")
plt.show()
```

```python
upper, middle, lower = talib.BBANDS(close, timeperiod=10, nbdevup=0.5, nbdevdn=0.5)
plt.figure(figsize=(11,3))
sma = talib.SMA(close)
plt.plot(close, "r-")
plt.plot(middle, "b-")
plt.fill_between(np.arange(len(close)), lower, upper, alpha=0.5)
plt.show()
```
