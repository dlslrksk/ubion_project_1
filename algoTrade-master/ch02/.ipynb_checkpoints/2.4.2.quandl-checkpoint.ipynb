```python
code_list = ["WIKI/AAPL","WIKI/AMZN","WIKI/MSFT"]
```

```python
data = quandl.get(dataset = code_list, rows=5)
```

```python
data.columns
```

```python
data
```

```python
extract_code = [x for x in data.columns if code_list[0] in x ]
```

```python
df = data.loc[:,extract_code]
```

```python
df.head()
```

```python
df.rename(columns=lambda x: x.replace('WIKI/AAPL - ', ''),inplace=True)
df
```

```python
for code in code_list:
    print(code)
    extract_code = [x for x in data.columns if code in x ]
    df = data.loc[:,extract_code]
    old_col = code+' - '
    df.rename(columns=lambda x: x.replace(old_col, ''),inplace=True)
    save_file = code.split('/')[1]+'.csv'
    df.to_csv(save_file)
```
