# 다수 종목 데이터 가져오기

#### KOSPI 가져오기

```python
result_df = pd.DataFrame() # 빈 프레임 만들기

for i in tqdm(range(1, 20)) :
    url = 'http://finance.naver.com/sise/field_submit.nhn'
    params = {
            'menu': 'market_sum',
            'returnUrl': 'http://finance.naver.com/sise/sise_market_sum.nhn?&page='+str(i),
            'fieldIds': ['quant','per','roe','pbr','dividend','market_sum', 'operating_profit_increasing_rate']
        }
    
    res = requests.get(url, params = params)
    df = pd.read_html(res.text)[1]
    df['배당수익률'] = (df['보통주배당금']/df['현재가'])*100
    df['시장'] = 'KOSPI'
    
    result_df = result_df.append(df[['종목명', '시가총액', '시장', 'PER', 'PBR', 'ROE', '배당수익률', '영업이익증가율']])
    
    
```

```python
result_df = result_df.dropna()
```

```python
result_df
```

```python
result_df.shape
```

#### 코스닥 더해서 추가하기

코스닥을 선택했을 경우 주소는?


```python
for i in tqdm(range(1, 28)) :
    url = 'http://finance.naver.com/sise/field_submit.nhn'
    params = {
            'menu': 'market_sum',
            'returnUrl': 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=1&page='+str(i),
            'fieldIds': ['quant','per','roe','pbr','dividend','market_sum', 'operating_profit_increasing_rate']
        }
    
    res = requests.get(url, params = params)
    df = pd.read_html(res.text)[1]
    df['배당수익률'] = (df['보통주배당금']/df['현재가'])*100
    df['시장'] = 'KOSDAQ'
    
    result_df = result_df.append(df[['종목명', '시가총액', '시장', 'PER', 'PBR', 'ROE', '배당수익률', '영업이익증가율']])
    
```
